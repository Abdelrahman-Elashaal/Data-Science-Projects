### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def correct_year(wrong_column):
    pattern=r'^20([2-9][4-9]|[3-9][0-9])'
    wrong_column=wrong_column.astype(str)
    wrong_column=wrong_column.str.replace(pattern,r'19\1',regex=True)
    wrong_column=pd.to_datetime(wrong_column)
    return wrong_column
```

```python
data.Yr_Mo_Dy=correct_year(data.Yr_Mo_Dy)
```
